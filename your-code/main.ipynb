# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# Your code here
# STEP 1 - Import the correct test from SciPy

from scipy.stats import ttest_1samp
```

```python
# step 2 - Filter hourly workers

hourly_workers = salaries[salaries['Salary or Hourly'] == 'Hourly']
```

```python
# STEP 3 - Extract hourly wages (drop missing values!)

hourly_wages = hourly_workers['Hourly Rate'].dropna()
```

```python
# STEP 4 - Perform the one-sample t-test

t_stat, p_value = ttest_1samp(hourly_wages, popmean=30)
```

```python
# STEP 5 - View results

t_stat, p_value
```

```python
# STEP 6 - Decision rule (α = 0.05)
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

A two-sided one-sample t-test was conducted to determine whether the mean hourly wage of hourly workers differs from $30 per hour.
The test yielded a t-statistic of 20.62 and a p-value of 4.32 × 10⁻⁹².
Since the p-value is far below the significance level of 0.05, we reject the null hypothesis.
This indicates that the mean hourly wage of hourly workers is significantly different from $30 per hour.

# Challenge 3 - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
# Your code here
# STEP 1 -  Import the required SciPy functions

from scipy.stats import t, sem

#t.interval → builds the confidence interval
#sem → computes the standard error of the mean
```

```python
# STEP 2 - Reuse the hourly wage data
# From Challenge 2 — no need to redo filtering

hourly_workers = salaries[salaries['Salary or Hourly'] == 'Hourly']
hourly_wages = hourly_workers['Hourly Rate'].dropna()
```

```python
# STEP 3 - Compute summary statistics

mean_wage = hourly_wages.mean()
std_error = sem(hourly_wages)
n = len(hourly_wages)
df = n - 1
```

```python
# STEP 4 - Construct the 95% confidence interval

confidence_interval = t.interval(
    confidence=0.95,
    df=df,
    loc=mean_wage,
    scale=std_error
)

confidence_interval
```

A 95% confidence interval for the mean hourly wage of hourly workers is [$32.52,\ $33.05].
This means we are 95% confident that the true population mean hourly wage lies between $32.52 and $33.05.

# Challenge 4 - Hypothesis Tests of Proportions

Another type of one sample test is a hypothesis test of proportions. In this test, we examine whether the proportion of a group in our sample is significantly different than a fraction. 

You can read more about one sample proportion tests [here](http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/SAS/SAS6-CategoricalData/SAS6-CategoricalData2.html).

In the cell below, use the `proportions_ztest` function from `statsmodels` to perform a hypothesis test that will determine whether the number of hourly workers in the City of Chicago is significantly different from 25% at the 95% confidence level.

```python
# Your code here

# Goal of the test: Determine whether the proportion of hourly workers in the City of Chicago is significantly different from 25%.

# Hypotheses (write this in your lab!)
# - Null hypothesis (H₀): p = 0.25
# - Alternative hypothesis (H₁): p ≠ 0.25
  # - (two-sided test, α = 0.05)

# STEP 1 - Import the correct function

from statsmodels.stats.proportion import proportions_ztest

```

Why z-test?

Large sample size
Proportion-based inference
Standard approach for one-sample proportion tests

```python
# STEP 2 - Define successes and sample size

# Here:
 # Success = employee is an hourly worker
 # Sample size = total number of employees

total_employees = len(salaries)
hourly_count = (salaries['Salary or Hourly'] == 'Hourly').sum()
```

```python
# STEP 3 - Perform the one-sample proportion z-test

z_stat, p_value = proportions_ztest(
    count=hourly_count,
    nobs=total_employees,
    value=0.25,
    alternative='two-sided'
)
```

```python
# STEP 4 - View results

z_stat, p_value
```

```python
# STEP 5 - Decision rule (α = 0.05)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

STEP 6 - INTERPRETATION

A one-sample proportion z-test was performed to determine whether the proportion of hourly workers in the City of Chicago is different from 25%.
The test yielded a z-statistic of −3.51 and a p-value of 0.00045.
